```python
import pandas as pd 
```

```python
path = "./"
var_df = pd.read_csv(path + "k_table.csv", encoding = 'cp949')
var_df.head(5)
```

```python
var_df['포화도'] = var_df['bpi']/var_df['기존충전소']
var_df.head(5)
```

```python
var_df_refine = var_df.drop(['기존충전소','주요건물수(영화관,대형점포,공원)','poplulation'], axis=1)
var_df_refine.head(5)
```

```python
bulid2_df = pd.read_csv(path + "건물2.csv", encoding = 'utf-8')
bulid3_df = pd.read_csv(path + "건물3.csv", encoding = 'utf-8')
```

```python
var_df_refine
```

```python
bulid3_df
```

```python
var_df = pd.merge(var_df_refine,bulid2_df, how = 'inner', on=['adng_nm'])
var_df = pd.merge(var_df,bulid3_df, how = 'inner', on=['adng_nm'])
var_df
```

```python
weight_vector = [0.2,0.1,0.4,0.6,0.4,0.1,0.2]
final_k_table = weight_vector * var_df.iloc[:,1:]
```

```python
final_k_table.to_csv("./k_table_final.csv", encoding = 'utf-8') 
```
