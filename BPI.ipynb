```python
import pandas as pd
```

```python
전기차보유대수 = pd.read_excel('BPI 준비(앱실행횟수, 전기차보유대수, 인구수).xlsx', sheet_name='전기차보유대수')
```

```python
앱실행횟수 = pd.read_excel('BPI 준비(앱실행횟수, 전기차보유대수, 인구수).xlsx', sheet_name='앱실행횟수')
```

```python
인구수 = pd.read_excel('BPI 준비(앱실행횟수, 전기차보유대수, 인구수).xlsx', sheet_name='인구수')
```

```python
k_table = pd.read_csv('k_table.csv', encoding='CP949')
```

```python
기존충전소 = k_table[['adng_nm', '기존충전소']]
```

```python
기존충전소.columns = ['읍면동', '기존충전소']
```

```python
기존충전소
```

```python
전기차보유대수['읍면동'] = 전기차보유대수['읍면동'].str.replace(' ','')
전기차보유대수['읍면동'] = 전기차보유대수['읍면동'].str.strip()
```

```python
전기차보유대수
```

```python
앱실행횟수.columns = ['읍면동', '앱실행횟수']
```

```python
앱실행횟수
```

```python
인구수['읍면동'] = 인구수['읍면동'].str.replace(' ','')
인구수['읍면동'] = 인구수['읍면동'].str.strip()
```

```python
인구수
```

```python
temp = pd.merge(전기차보유대수, 앱실행횟수, on='읍면동')
temp
```

```python
temp = pd.merge(temp, 인구수, on='읍면동')
temp
```

```python
temp['전기차보유대수비율'] = temp['전기차보유대수'] / temp['전기차보유대수'].sum()
temp['앱실행횟수비율'] = temp['앱실행횟수'] / temp['앱실행횟수'].sum()
temp['인구수비율'] = temp['인구수'] / temp['인구수'].sum()
```

```python
temp
```

```python
temp['구매력지수'] = 0.5 * temp['전기차보유대수비율'] + 0.3 * temp['앱실행횟수비율'] + 0.2 * temp['인구수비율']
```

```python
temp = pd.merge(temp, 기존충전소, on='읍면동')
temp
```

```python
temp['충전소포화도'] = temp['구매력지수'] / temp['기존충전소']
```

```python
temp[['읍면동','충전소포화도']].sort_values(by='충전소포화도', ascending=False)
```

```python
temp
```

```python
temp = pd.DataFrame(temp)
temp
```
